### Data Consolidation

```python
projects = pd.read_csv('../Data/Kaggle/KDDCup14/projects_feat.csv')
```

```python
outcomes = pd.read_csv('../Data/Kaggle/KDDCup14/outcomes_feat.csv')
```

```python
essays = pd.read_csv('../Data/Kaggle/KDDCup14/essays_feat.csv')
```

```python
resources = pd.read_csv('../Data/Kaggle/KDDCup14/resources_feat.csv')
```

```python
donations = pd.read_csv('../Data/Kaggle/KDDCup14/donations_feat.csv')
```

```python
df = projects.merge(outcomes)
```

```python
len(df)
```

```python
df = df.merge(essays)
```

```python
len(df)
```

```python
df = df.merge(resources, how='left')
```

```python
len(df)
```

```python
df = df.merge(donations)
```

```python
len(df)
```

```python
df.to_csv('../Data/Kaggle/KDDCup14/data_consolidation.csv', index=False)
```
